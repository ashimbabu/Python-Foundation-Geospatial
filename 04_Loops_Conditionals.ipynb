For Loops

```python
cities = ['San Francisco', 'Los Angeles', 'New York', 'Atlanta']

for city in cities:
    print(city)
```

```python
data = {'city': 'San Francisco', 'population': 881549, 'coordinates': (-122.4194, 37.7749) }

for x, y in data.items():
    print(x, y)
```

```python
for x in range(5):
    print(x)
```

```python
for x in range(1, 10, 2):
    print(x)
```

Conditionals

```python
for city in cities:
    if city == 'Atlanta':
        print(city)
```

```python
for city in cities:
    if city == 'Atlanta':
        print(city)
    else:
        print('This is not Atlanta')
```

```python
cities_population = {
    'San Francisco': 881549,
    'Los Angeles': 3792621,
    'New York': 8175133,
    'Atlanta':498044
}

for city, population in cities_population.items():
    if population < 1000000:
        print('{} is a small city'.format(city))
    elif population > 1000000 and population < 5000000:
        print('{} is a big city'.format(city))
    else:
        print('{} is a mega city'.format(city))
```

Control Statements

```python
for city in cities:
    print(city)
    if city == 'Los Angeles':
        print('I found Los Angeles')
        break
```

```python
for city in cities:
    if city == 'Los Angeles':
        continue
    print(city)
```

```python
for city in cities:
    if city == 'Los Angeles':
        pass
    else:
        print(city)
```

```python
for num in range(1, 100):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz", end=", ")
    elif num % 3 == 0:
        print("Fizz", end=", ")
    elif num % 5 == 0:
        print("Buzz", end=", ")
    else:
        print(num, end=", ")
```
