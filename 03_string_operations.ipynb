Printing Strings

```python
city = 'San Francisco'
population = 881549
output = 'Population of {} is {}.'.format(city, population)
print(output)
```

```python
latitude = 37.7749
longitude = -122.4194

coordinates = '{:.2f},{:.2f}'.format(latitude, longitude)
print(coordinates)
```
